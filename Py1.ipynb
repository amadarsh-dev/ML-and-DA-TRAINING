```python
# DATATYPES

# Integer (int): Represents whole numbers, positive or negative.
a = 12
print(f"Type of a: {type(a)}")

# String (str): Represents sequences of characters.
b = "yay"
print(f"Type of b: {type(b)}")

# Float (float): Represents real numbers with floating-point precision.
c = 10.00
print(f"Type of c: {type(c)}")

# Boolean (bool): Represents truth values (True or False).
d = True
print(f"Type of d: {type(d)}")

# List (list): An ordered, mutable collection of items.
my_list = [1, 2, 3, "hello"]
print(f"Type of my_list: {type(my_list)}")

# Tuple (tuple): An ordered, immutable collection of items.
my_tuple = (1, 2, "world")
print(f"Type of my_tuple: {type(my_tuple)}")

# Dictionary (dict): An unordered collection of key-value pairs.
my_dict = {"name": "Daisy"}
print(f"Type of my_dict: {type(my_dict)}")

# Set (set): An unordered collection of unique items.
my_set = {1, 2, 3, 3, 4}
print(f"Type of my_set: {type(my_set)}")
```

```python
# TYPECASTING - Changing data types!

# --- Implicit Typecasting (Automatic) ---
# Python changes types for you, like when adding int and float.
int_num = 10
float_num = 5.5
result_add = int_num + float_num # int_num became a float here!
print(f"Implicit (int + float): {int_num} + {float_num} = {result_add} (result type: {type(result_add).__name__})\n")

# --- Explicit Typecasting (Manual) ---
# We use functions like int(), float(), str() to change types.

# 1. To Int (int())
str_to_int = "123"
print(f"Str to Int: '{str_to_int}' becomes {int(str_to_int)}")
float_to_int = 10.75
print(f"Float to Int: {float_to_int} becomes {int(float_to_int)} (chops off decimal)\n")

# 2. To Float (float())
int_to_float = 50
print(f"Int to Float: {int_to_float} becomes {float(int_to_float)}")
str_to_float = "98.6"
print(f"Str to Float: '{str_to_float}' becomes {float(str_to_float)}\n")

# 3. To Str (str())
int_to_str = 100
print(f"Int to Str: {int_to_str} becomes '{str(int_to_str)}'")
bool_to_str = True
print(f"Bool to Str: {bool_to_str} becomes '{str(bool_to_str)}'\n")

# 4. To Bool (bool())
# 0, empty strings/lists, etc., are False. Others are True.
print(f"bool(0) is {bool(0)}")
print(f"bool(1) is {bool(1)}")
print(f"bool('') is {bool('')}")
print(f"bool('hello') is {bool('hello')}")
print(f"bool([]) is {bool([])}")
print(f"bool([1]) is {bool([1])}\n")

# 5. To List (list())
str_to_list = "Python"
print(f"Str to List: '{str_to_list}' becomes {list(str_to_list)}")
tuple_to_list = (1, 2, 3)
print(f"Tuple to List: {tuple_to_list} becomes {list(tuple_to_list)}\n")

# 6. To Tuple (tuple())
list_to_tuple = [4, 5, 6]
print(f"List to Tuple: {list_to_tuple} becomes {tuple(list_to_tuple)}")
str_to_tuple = "abc"
print(f"Str to Tuple: '{str_to_tuple}' becomes {tuple(str_to_tuple)}\n")

# 7. To Set (set())
list_to_set = [1, 2, 2, 3, 4]
# Sets remove duplicates!
print(f"List to Set: {list_to_set} becomes {set(list_to_set)}")
str_to_set = "banana"
print(f"Str to Set: '{str_to_set}' becomes {set(str_to_set)}\n")
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    return x / y

print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

while True:
    choice = input("Enter choice(1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter numbers only!")
            continue

        if choice == '1':
            print(f"{num1} + {num2} = {add(num1, num2)}")
        elif choice == '2':
            print(f"{num1} - {num2} = {subtract(num1, num2)}")
        elif choice == '3':
            print(f"{num1} * {num2} = {multiply(num1, num2)}")
        elif choice == '4':
            print(f"{num1} / {num2} = {divide(num1, num2)}")

        next_calculation = input("Let's do next calculation? (yes/no): ")
        if next_calculation.lower() == "no":
            break
    else:
        print("Invalid Input")
```
